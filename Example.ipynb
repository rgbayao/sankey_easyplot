```python
import pandas as pd
import seaborn as sns
import sankey_plot as sp
import requests
```

```python
url = 'https://data.mendeley.com/public-files/datasets/8gx2fvg2k6/files/72784be5-36d3-44fe-b75d-0edbf1999f65/file_downloaded'
r = requests.get(url, allow_redirects=True)
data = [x.split(',') for x in str(r.content.decode('latin1')).split('\n')]
df = pd.DataFrame(data[1:], columns = data[0])
```

```python
us_census = pd.read_csv('https://raw.githubusercontent.com/cphalpert/census-regions/master/us%20census%20bureau%20regions%20and%20divisions.csv')
```

```python
map_state_division = {j['State Code']:j['Division'] for i,j in us_census.iterrows()}
map_state_division['PR'] = 'Puerto Rico'
```

```python
df['Customer Division'] = df['Customer State'].apply(lambda x: map_state_division.get(x))
```

```python
sp.sankey_plot(df, 
            columns=['Customer Division', 
                    'Customer Segment', 
                    'Department Name',
                    'Market'],
            hue = 'Type',
            seaborn_palette='deep',
            title='Supply Chain Example'
)
```
